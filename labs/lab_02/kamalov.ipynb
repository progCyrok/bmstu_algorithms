# Сортировка Чёт-Нечёт
Этот относительно простой алгоритм сортировки, разработанный для использования на параллельных процессорах, является модификацией пузырьковой сортировки. 
Суть модификации в том, чтобы сравнивать элементы массива под чётными и нечётными индексами с последующими элементами независимо.

# Псевдокод

Функция OddEvenSort(массив):
    отсортирован = ЛОЖЬ                 # Изначально считаем, что массив не отсортирован
    ПОКА НЕ отсортирован:
        отсортирован = ИСТИНА           # Предполагаем, что массив отсортирован
       
        # Шаг 1: Четный проход
        ДЛЯ i ОТ 0 ДО длина(массив) - 2 С ШАГОМ 2:
            ЕСЛИ массив[i] > массив[i + 1]:
                Поменять местами массив[i] и массив[i + 1]   # Меняем элементы, если они в неправильном порядке
                отсортирован = ЛОЖЬ                          # Если были изменения, массив не отсортирован
        
        # Шаг 2: Нечетный проход
        ДЛЯ i ОТ 1 ДО длина(массив) - 2 С ШАГОМ 2:
            ЕСЛИ массив[i] > массив[i + 1]:
                Поменять местами массив[i] и массив[i + 1]   # Меняем элементы, если они в неправильном порядке
                отсортирован = ЛОЖЬ                          # Если были изменения, массив не отсортирован

    ВОЗВРАТИТЬ массив   # Когда массив отсортирован, возвращаем результат

# Сортировка Stooge Sort

**Stooge Sort** — это рекурсивный алгоритм сортировки, известный своей высокой временной сложностью. 
Он основан на сравнениях.
Алгоритм сначала проверяет первый элемент структуры данных и последний, обменивая их, если они находятся в неправильном порядке. 
Если в списке больше трёх элементов, он рекурсивно вызывает себя на начальных 2/3 списка,
на последних 2/3 списка и снова на начальных 2/3 списка, пока весь список не будет отсортирован.

# Блох-схема Сортировки STOOGE SORT
<img src="stooge_sort.png">

# Псевдокод
Функция StoogeSort(массив, левый, правый):
    ЕСЛИ массив[левый] > массив[правый] ТО:
        Поменять местами массив[левый] и массив[правый]

    ЕСЛИ (правый - левый + 1) > 2 ТО:  # Если в подмассиве больше 2 элементов
        m = ⌊(правый - левый + 1) / 3⌋  # Найти треть длины подмассива

        # Рекурсивно применить StoogeSort к первым 2/3 элементов массива
        StoogeSort(массив, левый, правый - m)

        # Рекурсивно применить StoogeSort к последним 2/3 элементам массива
        StoogeSort(массив, левый + m, правый)

        # Снова применить StoogeSort к первым 2/3 элементам массива
        StoogeSort(массив, левый, правый - m)

# Реализация Сортировки


```python
def stooge_sort(arr, start, end):
    if start >= end:
        return
    if arr[start] > arr[end]:
        arr[start], arr[end] = arr[end], arr[start]
    if end - start + 1 >= 3:
        third = (end - start + 1) // 3
        stooge_sort(arr, start, end - third)
        stooge_sort(arr, start + third, end)
        stooge_sort(arr, start, end - third)


arr = [7, 2, 1, 6, 5, 4, 3]
print(f"Изначальный массив: {arr}")
stooge_sort(arr, 0, len(arr) - 1)
print(f"Отсортированный массив: {arr}")
```

## Сравнение сортировки чёт-нечёт и сортировки Stooge Sort

**Сортировка чёт-нечёт (Odd-Even Sort):**

• Описание: Это сравнительный алгоритм сортировки, который работает, проходя по списку поочерёдно и сравнивая соседние элементы. Если два соседних элемента не в правильном порядке, они меняются местами. Процесс повторяется до тех пор, пока список не станет отсортированным.
• Сложность: 
  * Время: O(n^2) в худшем случае, O(n) в лучшем случае (когда список уже отсортирован)
  * Пространство: O(1) - сортировка происходит "на месте", не требуется дополнительной памяти.
• Преимущества:
  * Простой в реализации
  * Хорошо подходит для сортировки больших списков с небольшим количеством инверсий (неправильно упорядоченных пар)
• Недостатки:
  * Медленный по сравнению с более эффективными алгоритмами сортировки, такими как сортировка слиянием или быстрая сортировка
  * Не очень эффективен для списков с большим количеством инверсий.

**Сортировка Stooge Sort:**

• Описание: Рекурсивный алгоритм сортировки, который работает, разбивая список на три части и рекурсивно сортируя их.
• Сложность: 
  * Время: O(n^(log 3 / log 1.5)) - очень неэффективен, намного медленнее многих других алгоритмов сортировки
  * Пространство: O(n) - требуется дополнительная память для рекурсивных вызовов.
• Преимущества:
  * Интересная реализация рекурсии
  * Может быть понятна и использована в качестве учебного примера 
• Недостатки:
  * Крайне неэффективен в реальных сценариях, его время работы очень высокое
  * Сложно реализовать и понять для начинающих

**Вывод:**

- Сортировка чёт-нечёт - простой, эффективный алгоритм сортировки, который можно использовать в случаях, когда требуется небольшая сложность реализации и сортируются большие списки с небольшим количеством инверсий. 
- Сортировка Stooge Sort - непрактичный алгоритм, который больше подходит для учебных целей. В реальных сценариях использование Stooge Sort не рекомендуется из-за её очень большой временной сложности.

**Рекомендации:**

* Для большинства задач сортировки используйте более эффективные алгоритмы, такие как сортировка слиянием или быстрая сортировка.
* Сортировка чёт-нечёт может быть полезна для сортировки больших массивов с небольшим количеством инверсий.
* Stooge Sort можно использовать в учебных целях для демонстрации рекурсии, но не для практического применения.
